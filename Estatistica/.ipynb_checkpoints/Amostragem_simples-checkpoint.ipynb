```python
# Mudando a semente aleatória para que sempre que for executado seja sempre o mesmo resultado
np.random.seed(2345)
# 150 amostras, de 0 a 1, com respoição, probabilidades equivalentes
#np.randow.choice(espaço_amostral(de vai gerar a amostra), tamanho, com_reposição, probabilidade(de 70% para gerar 0 e 30% para gerar 1))
amostra = np.random.choice(a=[0,1], size=150, replace=True, p=[0.7, 0.3])

# Verificar o tamanho da amostra
len(amostra)
# Verificar o tamanho da amostra para valores iguais a 0 e 1
len(amostra[amostra==1])
len(amostra[amostra==0])
amostra
```
