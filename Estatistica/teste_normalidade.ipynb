# Teste de Normalidadde

```python
# Importação das bibliotecas: matplotlib para geração de gráficos
# norm - gera dados normalmente distribuidos
# skewnorm - gera dados com inviesamento
from scipy import stats
from scipy.stats import norm, skewnorm  
import matplotlib.pyplot as plt
```

## Dados Normalmente Distribuídos

```python
# Criação de uma variável com dados em uma distribuição normal com a função rvs (1000 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
# Geração do histograma
plt.hist(dados, bins = 20) # bins - número de eixos(barras)
plt.title('Dados')
```

```python
# Geração do gráfico para verificar se estão normalmente distribuídos
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show
```

```python
# Execução do teste de Shapiro
# Segundo argumento é o valor de p, não há como rejeitar a hipotese nula, hipótese nula o valor deve ser em torno de 0,05
stats.shapiro(dados)
```

## Dados Inviesados

```python
# Geração de dados inviesados
dados2 = skewnorm.rvs(4, size = 1000)
```

```python
# Geração do histograma
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
# Geração do gráfico para verificar se estão normalmente distribuídos
fig, ax = plt.subplots()
stats.probplot(dados2, fit = True, plot = ax)
plt.show
```

```python
# Execução do teste de Shapiro
# Segundo argumento é o valor de p, não há como rejeitar a hipotese nula, hipótese nula o valor deve ser em torno de 0,05
stats.shapiro(dados2)
```
