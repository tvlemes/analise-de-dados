# Grafico Densidade

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.shape
```

```python
# Histograma com 10 divisoes somente para o primeiro atributo da base de dados
plt.hist(base.iloc[:,1], bins = 6)
```

```python
# OBS: Grafico depreciado, deve ser usado o displot
# Histograma com a linha de distribuicao de frequencia, com 6 divisoes(bins)
# kde - linha de densidade
sns.distplot(base.iloc[:,1], 
            hist = True,
            kde = False,
           bins = 6, color = 'blue',
          hist_kws = {'edgecolor': 'black'})
```

```python
# Mesmo Histograma acima mas utilizando o grafico displot
sns.displot(base.iloc[:,1], 
            kde = False,
           bins = 6, color = 'blue')
```

```python
# OBS: Grafico depreciado, deve ser usado o displot
# Densidade
sns.distplot(base.iloc[:,1], 
            hist = False,
            kde = True,
           bins = 6, color = 'blue',
          hist_kws = {'edgecolor': 'black'})
```

```python
# Mesmo Histograma acima mas utilizando o grafico displot
sns.displot(base.iloc[:,1], 
            kind="kde", 
            color = 'blue')
```

```python
# OBS: Grafico depreciado, deve ser usado o displot
# Densidade e Histograma
sns.distplot(base.iloc[:,1], 
            hist = True,
            kde = True,
           bins = 6, color = 'blue',
          hist_kws = {'edgecolor': 'black'})
```

```python
# Mesmo Histograma acima mas utilizando o grafico displot
sns.displot(base.iloc[:,1], 
            kde = True,
           bins = 6, color = 'blue')
```
